```python
import pandas as pd
```

```python
election_df = pd.DataFrame(columns=['election_date','election_type','foregin_registered_voters','foregin_voted_voters','total_voters','total_voted_voters','who_win'])
election_df.head()
```

```python
pres_2005_I = pd.read_csv(r"C:\Users\kgrab\Desktop\Python Jupyte Notebook\historical_election_poland_da\data_source\presidential\pres_2005_I.csv",sep=';')
```

```python
pres_2005_I.head()
```

```python
pres_2005_I['Uprawnieni do głosowania'] = pres_2005_I['Uprawnieni do głosowania'].str.replace(' ','').astype(int)
pres_2005_I['Karty wydane (frekwencja)'] = pres_2005_I['Karty wydane (frekwencja)'].str.replace(' ','').astype(int)
pres_2005_I.info()
```

```python
new_row = pd.DataFrame([
                    {'election_date':'2005-10-09',
                    'election_type':'pres_I',
                    'foregin_registered_voters':pres_2005_I['Uprawnieni do głosowania'].where(pres_2005_I['Powiat']=='Zagranica').sum(),
                    'foregin_voted_voters':pres_2005_I['Karty wydane (frekwencja)'].where(pres_2005_I['Powiat']=='Zagranica').sum(),
                    'total_voters':pres_2005_I['Uprawnieni do głosowania'].sum(),
                    'total_voted_voters':pres_2005_I['Karty wydane (frekwencja)'].sum(),
                    'who_win':'PO'}
])

election_df = pd.concat([election_df,new_row], ignore_index=True)

election_df.head()
```
